```python
movie_rate_count=netflix_dataset['movie_id'].value_counts()
```

```python
movie_rate_count # these are the unique no. of ratings a movie got.
```

```python
# we need movies above 60% rating threshold.
threshold=round(movie_rate_count.quantile(0.6))
threshold
```

```python
dropped_movies=movie_rate_count[movie_rate_count < threshold].index #these are the movies which are less than threshold review.
dropped_movies
```

```python
# if we find the number of customers then we can find that how many ratings each people has done.
customer_review_count=netflix_dataset['customer_id'].value_counts()
customer_review_count
```

```python
# Benchmark for the customer reviw:-
bench_mark=customer_review_count.quantile(0.6)
bench_mark
```

```python
# dropped customers:
dropped_customer=customer_review_count[customer_review_count < bench_mark].index
dropped_customer
```

```python
netflix_dataset.info()
```

```python
# now we need these customers and movies which are not in dropped.using isin() method we will perform.
netflix_dataset=netflix_dataset[~netflix_dataset['customer_id'].isin(dropped_customer)]
netflix_dataset=netflix_dataset[~netflix_dataset['movie_id'].isin(dropped_movies)]
```

```python
# =============================
# 5. RECOMMENDATION FUNCTION
# (Enhanced Visual Output)
# =============================

def recommend_movies(user_id):

    user_id = int(user_id)

    movies_copy = movie_title.copy()

    movies_copy['Predicted Rating'] = movies_copy['movie_id'].apply(
        lambda x: model.predict(user_id, x).est
    )

    top10 = movies_copy.sort_values(
        by="Predicted Rating",
        ascending=False
    ).head(10).reset_index(drop=True)


    # Create Netflix Style Cards
    cards = ""

    for i, row in top10.iterrows():

        rating_percent = (row['Predicted Rating'] / 5) * 100

        cards += f"""

        <div class="movie-card">

            <div class="rank">#{i+1}</div>

            <div class="movie-name">{row['name']}</div>

            <div class="rating-bar-bg">

                <div class="rating-bar-fill"
                     style="width:{rating_percent}%">

                     ⭐ {row['Predicted Rating']:.2f}

                </div>

            </div>

        </div>

        """

    return f"<div class='card-container'>{cards}</div>"



# =============================
# 6. ULTRA NETFLIX CSS
# =============================

css = """

/* ===== GLOBAL BACKGROUND ===== */

body {

    background-color: #000000 !important;

}


/* ===== MAIN CONTAINER ===== */

.gradio-container {

    background-color: #000000 !important;

}


/* ===== NETFLIX AI GLOWING HEADER ===== */

#title {

text-align:center;
font-size:60px;
font-weight:900;
color:#E50914;
margin-top:20px;
margin-bottom:10px;
letter-spacing:4px;

/* GLOW EFFECT */

text-shadow:

0 0 10px #E50914,
0 0 20px #E50914,
0 0 40px #E50914,
0 0 80px #E50914;

animation: glowPulse 2s infinite alternate;

}


/* Glow animation */

@keyframes glowPulse {

from {

text-shadow:
0 0 10px #E50914,
0 0 20px #E50914;

}

to {

text-shadow:
0 0 20px #ff1a1a,
0 0 40px #ff1a1a,
0 0 80px #ff1a1a;

}

}


/* ===== SUBTITLE ===== */

#subtitle {

text-align:center;
font-size:22px;
color:white;
margin-bottom:40px;

}


/* ===== DROPDOWN ===== */

.gr-dropdown {

background-color:#111111 !important;
color:white !important;
border:1px solid #E50914 !important;

}


/* ===== BUTTON ===== */

.gr-button {

background: linear-gradient(90deg, #E50914, #ff0000) !important;

color:white !important;

font-size:18px !important;

font-weight:bold;

border:none !important;

border-radius:10px !important;

padding:15px !important;

box-shadow:0 0 20px #E50914;

transition:0.3s;

}


.gr-button:hover {

transform: scale(1.08);

box-shadow:

0 0 20px #ff0000,
0 0 40px #ff0000,
0 0 60px #ff0000;

}


/* ===== CARD CONTAINER ===== */

.card-container {

display:flex;
flex-wrap:wrap;
justify-content:center;
gap:30px;

}


/* ===== MOVIE CARD WITH RED GLOW ===== */

.movie-card {

background: linear-gradient(145deg, #0d0d0d, #1a0000);

padding:25px;

border-radius:15px;

width:420px;

border:1px solid #E50914;

box-shadow:

0 0 10px #E50914,
0 0 20px #E50914;

transition:0.4s;

}


/* HOVER GLOW */

.movie-card:hover {

transform: scale(1.07);

box-shadow:

0 0 20px #ff0000,
0 0 40px #ff0000,
0 0 80px #ff0000;

}


/* ===== RANK ===== */

.rank {

color:#ff1a1a;

font-size:30px;

font-weight:bold;

margin-bottom:10px;

}


/* ===== MOVIE NAME ===== */

.movie-name {

color:white;

font-size:22px;

font-weight:600;

margin-bottom:15px;

}


/* ===== RATING BAR ===== */

.rating-bar-bg {

background:#330000;

border-radius:20px;

}


.rating-bar-fill {

background: linear-gradient(90deg, #ff0000, #E50914);

padding:10px;

border-radius:20px;

color:white;

font-weight:bold;

box-shadow:0 0 10px #ff0000;

}

"""



# =============================
# 7. BUILD NETFLIX PRO UI
# =============================

import gradio as gr

user_list = sorted(netflix_dataset['customer_id'].unique())


with gr.Blocks(css=css, theme=gr.themes.Soft()) as app:


    gr.HTML("<div id='title'>NETFLIX AI</div>")

    gr.HTML("<div id='subtitle'>Personalized Movie Recommendation Engine</div>")


    with gr.Row():

        user_dropdown = gr.Dropdown(

            choices=user_list,
            label=" Select User ID",
            value=user_list[0],
            interactive=True

        )


    recommend_btn = gr.Button(" Recommend Movies")


    output_cards = gr.HTML()



    def display(user):

        return recommend_movies(user)



    recommend_btn.click(

        fn=display,
        inputs=user_dropdown,
        outputs=output_cards

    )



# =============================
# 8. LAUNCH
# =============================

app.launch()
```
