# CUSTOMER ID COLUMN CONSIST OF BOTH MOVIE ID AND CUSTOMER ID,WHICH IS REPRESENTED AS NAN AT RATINGS COLUMN.

```python
# NUMBER OF MOVIES:
movie_count=netflix_dataset.isnull().sum()
movie_count=movie_count['ratings']
movie_count
```

```python
# FINDING OF TOTAL NUMBER OF CUSTOMERS:-
total_customer=netflix_dataset['customer_id'].nunique() - movie_count
```

```python
total_customer #TOTAL NUMBER OF CUSTOMERS
```

```python
# COUNT OF RATINGS:-
rating_count=netflix_dataset['ratings'].count() - movie_count
```

```python
rating_count
```

```python
given_ratings=netflix_dataset['ratings'].value_counts() #unique count of ratings
```

```python
# NOW WE HAVE TO REMOVE THE NUMBER OF MOVIE ID FROM CUSTOMER ID COL,AND NEED TO SEPERATE THEM.
customer_ids=[]
for value in netflix_dataset['customer_id']:
  if ':' in value:
    id=value.replace(':','')
  customer_ids.append(id)
```

```python
customer_ids
```

```python
netflix_dataset['movie_id']=customer_ids #ADDING THE ARRAY WITH THE DATASET.
```

```python
netflix_dataset
```

```python
# now removing the cust_id from custimer_if column.
netflix_dataset.dropna(inplace=True)
```

```python
netflix_dataset
```

```python
netflix_dataset.info()
```

```python
# CONVERTING THE CUSTOMER_ID INTO INT
netflix_dataset['customer_id']=netflix_dataset['customer_id'].astype(int)
```

```python
netflix_dataset.info()
```

```python
movie_rate_count=netflix_dataset['movie_id'].value_counts()
```

```python
movie_rate_count # these are the unique no. of ratings a movie got.
```

```python
# we need movies above 60% threshold.
threshold=round(movie_rate_count.quantile(0.6))
threshold
```

```python
dropped_movies=movie_rate_count[movie_rate_count<threshold].index #these are the movies which are less than threshold review.
dropped_movies
```

```python
# if we find the number of customers then we can find that how many ratings each people has done.
customer_review_count=netflix_dataset['customer_id'].value_counts()
customer_review_count
```

```python
# Benchmark for the customer reviw:-
bench_mark=customer_review_count.quantile(0.6)
bench_mark
```

```python
# dropped customers:
dropped_customer=customer_review_count[customer_review_count < bench_mark].index
dropped_customer
```

```python
netflix_dataset.info()
```

```python
# now we need these customers and movies which are not in dropped.using isin() method we will perform.
netflix_dataset=netflix_dataset[~netflix_dataset['customer_id'].isin(dropped_customer)]
netflix_dataset=netflix_dataset[~netflix_dataset['movie_id'].isin(dropped_movies)]
```

```python
#importing libraries:
from surprise import SVD,Reader,Dataset #svd:-singular value decomposition.
from surprise.model_selection import cross_validate
```

```python
reader=Reader() #its for reading the objects
```

```python
data=Dataset.load_from_df(netflix_dataset[['movie_id','customer_id','ratings']][:1000000],reader)
```

```python
data #data is stored into the object
```

```python
model=SVD()
```

```python
#cross validation technique using rmse(root mean square error)
cross_validate(model,data,measures=['RMSE'],cv=3)
```
